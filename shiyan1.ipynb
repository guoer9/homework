```python
#使用函数生成描述性统计，总结数据集分布的中心趋势，分散和形状。
from scipy.stats import skew, kurtosis
# 中心趋势: 计算均值和中位数
mean = df.mean().mean()  # 所有月份的均值
median = df.median().median()  # 中位数

# 分散程度: 标准差、极差和四分位间距
std_dev = df.std().mean()  # 标准差
range_ = df.max().max() - df.min().min()  # 极差
iqr = df.quantile(0.75).mean() - df.quantile(0.25).mean()  # 四分位间距

# 形状: 偏度和峰度
skewness = skew(df.stack())  # 偏度
kurt_value = kurtosis(df.stack())  # 峰度

# 输出结果
print(f"中心趋势: \n均值: {mean}, 中位数: {median}")
print(f"\n分散程度: \n标准差: {std_dev}, 极差: {range_}, 四分位间距: {iqr}")
print(f"\n分布形状: \n偏度: {skewness}, 峰度: {kurt_value}")
```

```python
#求增长率
df['sum']=year_sum
df['累计']=1.0 * df['sum'].cumsum()
df['增长率']= df['累计']/df['sum'].sum()*100
df
```

```python
#使用帕累托图表达出航空乘客数量与年份的关系
fig = plt.figure(figsize=(10,6))
ax1= fig.add_subplot(111)
ax1.bar(data['year'], year_sum, color='skyblue')
ax2 = ax1.twinx()
ax2.plot(data['year'], df['增长率'], color='red',marker='o')
plt.title('每年乘客变化趋势图')
plt.ylabel('增长率')
plt.xlabel('年份')
plt.show()
```

```python
#分析周期性变化
from statsmodels.tsa.seasonal import seasonal_decompose
df3=df.stack().reset_index(level=1, drop=True)
df3 = seasonal_decompose(df3,model='multiplicative', period=12)
df3.plot()
plt.show()
```
